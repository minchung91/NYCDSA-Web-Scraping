```python
import pandas as pd
df1 = pd.read_csv('df1.csv')
df1
```

```python
a = df1.rename(columns={'2015 Honda Civic LX': 'yearmakemodel'})
b = a.rename(columns={'Price: $9,980$165/mo est.': 'Price'})
c = b.rename(columns={'25,704': 'Mileage'})
d = c.rename(columns={'Black': 'Color'})
e = d.rename(columns={'Sedan': 'Body'})
f = e[~e.Price.str.contains("Location:")]
```

```python
print(f["Price"])
h = f["Price"].tolist()
j = [i.replace('Price: $', '') for i in h]
j
```

```python
k = [i.split('$', 1)[0] for i in j]
k
```

```python
l = [s.replace(',', '') for s in k]
l
```

```python
m = pd.Series(l)
f['Price'] = m.values
```

```python
f
```

```python
mil1 = f["Mileage"].tolist()
mil2 = [s.replace(',', '') for s in mil1]
mil3 = pd.Series(mil2)
f['Mileage'] = mil3.values
f
```

```python
ymm = f["yearmakemodel"].tolist()
ymm1 = [i.split(' ', 1)[0] for i in ymm]
ymm1
# mil2 = [s.replace(',', '') for s in ymm]
year = pd.Series(ymm1)
f['Year'] = year.values
f
```

```python
ymm = [e[5:] for e in ymm]
```

```python
make = [i.split()[0] for i in ymm]
make = pd.Series(make)
f['Make'] = make.values
f
```

```python
import numpy as np
import matplotlib.pyplot as plt
y = chart2["Mileage"].tolist()
s = chart2["Year"].tolist()
p = chart2["Price"].tolist()
ya = list(map(int, y))
sa = list(map(int, s))
pa = list(map(int, p))
yb = ya / np.max(ya,0)
sb = sa / np.max(sa,0)
pb = (pa / np.max(pa,0))*900
N = 42
colors = np.random.rand(N)
plt.scatter(sb, yb, pb, c=colors, alpha=0.5)
plt.ylim(ymax = 1, ymin = 0)
plt.xlim(xmax = 1.005, xmin = 0.99)
plt.xlabel("Relative Year")
plt.ylabel("Relative Mileage")
plt.title("Relative year, mileage, and price")
plt.show()
```
